# Exploratory Data Analysis

# Gender


```python
# plotting a bar chart for Gender and it's count


# Create a countplot for the 'Gender' column
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Gender', data=Diwali1)

# Calculate the total number of entries and add labels with count and percentage
total = len(Diwali1)
for container in ax.containers:
    ax.bar_label(container, labels=[f'{int(v)}\n{v/total:.1%}' for v in container.datavalues])

# Add labels and title
ax.set_title('Count of Genders with Percentage')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')

# Display the plot
plt.show()
```

```python
AM_gen = Diwali1.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
AM_gen
```

```python
sns.barplot(x='Gender',y='Amount',data=AM_gen)
plt.show()
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = Diwali1.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Most of the buyers belong to which age group?¶

```python


# Create a countplot with 'Age Group' on the x-axis and 'Gender' as the hue
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=Diwali1, x='Age Group', hue='Gender')

# Calculate total counts per group for percentage calculation
total_counts = Diwali1.groupby('Age Group').size()

# Add labels with count and percentage
for container in ax.containers:
    labels = []
    for i, bar in enumerate(container):
        count = int(bar.get_height())
        total = total_counts[i]
        percentage = f'{count / total:.1%}'
        labels.append(f'{count}\n{percentage}')
    ax.bar_label(container, labels=labels)

# Add labels and title
ax.set_title('Count of Age Groups by Gender with Percentage')
ax.set_xlabel('Age Group')
ax.set_ylabel('Count')

# Display the plot
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = Diwali1.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### Most of the orders are from which state?

```python
# total number of orders from top 10 states

sales_state = Diwali1.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
plt.xticks(rotation=45)
```

```python
# total amount/sales from top 10 states

sales_state = Diwali1.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
plt.xticks(rotation=45)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Is there any relationship between Marital Status and purchasing behaviour?

The loop iterates over each container of bars in the plot.
For each bar, it calculates both the count (int(v)) and
the percentage (v/total), where v is the height of the bar (i.e., the count of the category).
Inside the loop, labels are created that include both the count and the percentage. The percentage is formatted to show one decimal place using .1%.

```python


# Set the figure size
sns.set(rc={'figure.figsize': (7, 5)})

# Create a countplot for the 'Marital_Status' column
ax = sns.countplot(data=Diwali1, x='Marital_Status')

# Calculate the total number of entries for percentage calculation
total = len(Diwali1)

# Add labels with count and percentage
for container in ax.containers:
    labels = [f'{int(v)}\n{v/total:.1%}' for v in container.datavalues]
    ax.bar_label(container, labels=labels)

# Add labels and title
ax.set_title('Count of Marital Status with Percentage')
ax.set_xlabel('Marital Status')
ax.set_ylabel('Count')

# Display the plot
plt.show()
```

```python
sales_state = Diwali1.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### What is the Occupation of most of the buyers?

```python

# Set the figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Create a countplot for the 'Occupation' column
ax = sns.countplot(data=Diwali1, x='Occupation')

# Calculate the total number of entries for percentage calculation
total = len(Diwali1)

# Add labels with count and percentage
for container in ax.containers:
    labels = [f'{int(v)}\n{v/total:.1%}' for v in container.datavalues]
    ax.bar_label(container, labels=labels)

# Add labels and title (optional)
ax.set_title('Count of Occupations with Percentage')
ax.set_xlabel('Occupation')
ax.set_ylabel('Count')

# Display the plot
plt.show()
```

```python
sales_state = Diwali1.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Which Product Category has drawn maximum sales?¶


```python
plt.figure(figsize=(10,5))
Total_sales_Orders = Diwali1.groupby('Product_Category')['Orders'].sum().reset_index()
cx=sns.barplot(x='Product_Category',y='Orders',data = Total_sales_Orders.nlargest(10, 'Orders'))
cx.set(title="Total Orders(Top 10) Vs Product Categories")
for bars in cx.containers:
    cx.bar_label(bars)
plt.xticks(rotation=45)
```

```python


# Set the figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Create a countplot for the 'Product_Category' column
ax = sns.countplot(data=Diwali1, x='Product_Category')

# Calculate the total number of entries for percentage calculation
total = len(Diwali1)

# Add labels with count and percentage
for container in ax.containers:
    labels = [f'{int(v)}\n{v/total:.1%}' for v in container.datavalues]
    ax.bar_label(container, labels=labels)

# Add labels and title (optional)
ax.set_title('Count of Product Categories with Percentage')
ax.set_xlabel('Product Category')
ax.set_ylabel('Count')

# Display the plot
plt.show()
```

```python
sales_state = Diwali1.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
